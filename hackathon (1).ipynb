```python
#splitting data
X = train[features].copy()
X_test = train[features].copy()


from sklearn.model_selection import train_test_split
# Break off validation set from training data
X_train, X_valid, z_train, z_valid = train_test_split(X, z,
                                                      random_state=100,train_size=1, test_size=1)
```

```python
#fitting the model
from sklearn.ensemble import RandomForestRegressor
my_model =RandomForestRegressor(bootstrap=True, ccp_alpha=0.0, criterion='mse',
                      max_depth=5, max_features='auto', max_leaf_nodes=None,
                      max_samples=1, min_impurity_decrease=0.0,
                      min_impurity_split=None, min_samples_leaf=1,
                      min_samples_split=2, min_weight_fraction_leaf=0.0,
                      n_estimators=25, n_jobs=2, oob_score=False,
                      random_state=100, verbose=0, warm_start=False)
my_model.fit(X, z)
```

```python
#predicting the out come
preds_test = my_model.predict(X_test)
print(preds_test)
```
